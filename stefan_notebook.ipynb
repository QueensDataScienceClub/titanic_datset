```python
import pandas as pd
import sklearn as sk
#from sklearn import linear_model
```

```python
train_df = pd.read_csv('datasets/train.csv')
test_df = pd.read_csv('datasets/test.csv')
test_target_df = pd.read_csv('datasets/gender_submission.csv')
```

```python
test_df = pd.concat([test_df, test_target_df], axis=1)
```

```python
# Drop nulls
train_df.dropna(inplace=True)
test_df.dropna(inplace=True)
```

```python
# Set a target dataframe
train_target_df = train_df['Survived']
test_target_df = test_df['Survived']
```

```python
# Drop the Rows that we dont want
train_df.drop(['Survived', 'Name', 'Ticket', 'Cabin', 'Embarked', 'PassengerId'], inplace=True, axis=1)
test_df.drop(['Survived', 'Name', 'Ticket', 'Cabin', 'Embarked', 'PassengerId'], inplace=True, axis=1)
```

```python
# One hot encode the Sex cloumn
train_df['Male'] = (train_df['Sex'] == 'male').astype(int)
train_df['Female'] = (train_df['Sex'] == 'female').astype(int)
train_df.drop(['Sex'], inplace=True, axis=1)

test_df['Male'] = (test_df['Sex'] == 'male').astype(int)
test_df['Female'] = (test_df['Sex'] == 'female').astype(int)
test_df.drop(['Sex'], inplace=True, axis=1)
```

```python
# fit to a Logistic Regression model
model = sk.linear_model.LogisticRegression()
model.fit(train_df, train_target_df)
y_pred = model.predict(test_df)
```

```python
# Checking the accuracy
sk.metrics.accuracy_score(test_target_df, y_pred)
#model.score(test_target_df, y_pred)
```
